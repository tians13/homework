```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel('学生成绩单.xls',header=0)
```

```python
df = data.drop(['Unnamed: 8','Unnamed: 9'],axis=1)[:68]
```

```python
df
```

```python
total_data = df[['语文','数学','英语']].astype(float)
```

```python
mu = total_data.mean().values
sigma = np.cov(total_data.T)
print("均值向量 μ：", mu)
print("协方差矩阵 Σ：\n", sigma)
```

```python
print(f'语文均值：{mu[0]},方差：{df['语文'].var()}')
print(f'数学均值：{mu[1]},方差：{df['数学'].var()}')
print(f'英语均值：{mu[2]},方差：{df['英语'].var()}')
```

```python
print(f'多维高斯模型的均值向量：{mu} \n协方差矩阵:\n{sigma}')
```

```python
# 计算相关系数矩阵 ρ
corr_matrix = np.corrcoef(total_data.T)
print("协方差矩阵:")
print(sigma)
print('相关系数矩阵：')
print(corr_matrix)
```

```python
# 生成满足多维高斯分布的 100 组三科成绩
generated_scores = np.random.multivariate_normal(mu, sigma, 100)

# 转换为 DataFrame 便于查看
generated_scores_df = pd.DataFrame(generated_scores, columns=['语文', '数学', '英语'])
generated_scores_df.to_excel('generated_scores.xlsx', index=False)
```

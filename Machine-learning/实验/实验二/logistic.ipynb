模型定义

```python
def loss(y, y_hat):
    y_hat = np.clip(y_hat, 1e-15, 1 - 1e-15)
    m = len(y)
    loss = -(1 / m) * np.sum(y * np.log(y_hat) + (1 - y) * np.log(1 - y_hat))
    return loss
def linear(X,w,b):
    return X @ w + b
def gradient_descent(X,y,y_hat,lr):
    global w,b
    m = len(X)
    l = (y_hat-y.reshape(y_hat.shape)).reshape(-1,1)
    w_gradient = X.T @ l/m
    b_gradient = np.sum(l)/m
    w -= lr*w_gradient
    b -= lr*b_gradient
def logistic(z):
    return 1/(1+np.exp(-z))
def data_iter(batch_size,X,y):
    num = len(X)
    index = np.arange(num)
    np.random.shuffle(index)
    for i in range(0,num,batch_size):
        j = np.array(index[i:min(i+batch_size,num)])
        yield X[j],y[j]
def train(X,y,lr,epochs,batch_size):
    for epoch in range(epochs):
        for X_train,y_train in data_iter(batch_size,X,y):
            y_hat = logistic(linear(X_train,w,b))
            train_loss = loss(y_train,y_hat)
            gradient_descent(X_train,y_train,y_hat,lr)
        print(f'epoch {epoch + 1}, loss {float(train_loss.mean()):f}')
```

定义超参数

```python
w = np.ones((X_train.shape[1],1))
b = 0
lr = 0.1
epochs = 1000
batch_size = 1024
```

```python
train(X_train,y_train,lr,epochs,batch_size)
```

计算准确率

```python
y_pred = logistic(linear(X_test,w,b))
y_pred[y_pred>=0.5] = 1
y_pred[y_pred<0.5] = 0
accuracy = np.sum(y_pred==y_test)/len(y_test)
print(f'准确率:{accuracy}')
```

计算精确率

```python
# 预测为正类的样本数量
TP = np.sum(np.logical_and(y_pred == 1, y_test == 1))
# 被预测为正类的样本数量
FP = np.sum(y_pred == 1) - TP

# 精确率 = TP / (TP + FP)
precision = TP / (TP + FP)
print(f"精确率: {precision}")
```

```python
# 计算召回率
# 预测为正类的样本数量
TP = np.sum(np.logical_and(y_pred == 1, y_test == 1))
# 实际为正类的样本数量
TP_plus_FN = np.sum(y_test == 1)

# 召回率 = TP / TP_plus_FN
recall = TP / TP_plus_FN
print(f"召回率: {recall}")
```

```python
from sklearn.metrics import roc_curve, auc
# 计算 ROC-AUC
fpr, tpr, _ = roc_curve(y_test, y_pred)
roc_auc = auc(fpr, tpr)

# 输出 ROC-AUC 值
print(f"ROC-AUC 值: {roc_auc}")
```
